## Подключение бибилиотек

```python
# Работа с данными
import pandas as pd
import numpy as np

# Библиотеки для визуализации
import seaborn as sns
import matplotlib.pyplot as plt
```

## Загрузка данных

```python
# Начнем с чтения csv данных
df = pd.read_csv('sample_data/spotify-2023_fixed.csv')
```

```python
# Первые пять строк
df.head()
```

```python
# Создание новой фигуры с установкой размеров
plt.figure(figsize=(26, 12))

sns.scatterplot(data=df,
                x='in_spotify_charts',
                y='released_year',
                style="artist_count")

# Добавление подписей осей
plt.xlabel("Присутствие и рейтинг песни в чартах Spotify")
plt.ylabel("Год выпуска")
```

```python
import seaborn.objects as so
import matplotlib as mpl

f = mpl.figure.Figure(figsize=(26, 8))
(
    so.Plot(df, y="energy_%", color="speechiness_%")
    .pair(x=["in_spotify_charts", "in_apple_charts", "in_deezer_charts"])
    .facet(row="mode")
    .on(f)
    .add(so.Dots())
)
```

```python
sns.displot(df, x="released_month", hue="artist_count", element="step")

# Добавление подписей осей
plt.xlabel("Месяц выпуска")
plt.ylabel("Кол-во выпущеных песен")
```

```python
# Установка размера графика для всех последующих графиков
sns.set(rc={'figure.figsize': (16, 8)})

sns.scatterplot(data=df, x='energy_%', y='speechiness_%', hue='mode')

plt.legend(title='Режим песни')

# Добавление подписей осей
plt.xlabel("Воспринимаемый уровень энергии песни")
plt.ylabel("Кол-во произнесенных слов в песне")
```

```python
sns.scatterplot(data=df,
                x='in_spotify_playlists',
                y='in_apple_playlists',
                hue='speechiness_%',
                palette="flare")

plt.legend(title='Кол-во слов в песне')

# Добавление подписей осей
plt.xlabel("Кол-во плейлистов Spotify, в которые песня включена")
plt.ylabel("Кол-во плейлистов Apple Music, в которые песня включена")
```

```python
sns.scatterplot(data=df,
                x='in_spotify_playlists',
                y='danceability_%',
                hue='released_year',
                palette="flare")

plt.legend(title='Год выпуска')

# Добавление подписей осей
plt.xlabel("Кол-во плейлистов Spotify, в которые песня включена")
plt.ylabel("% насколько песня подходит для танцев")
```
